```python
import pandas as pd
#This data desmonstrated the ZHVI of Single Family Housing in the msa
df = pd.read_csv("ZHVI_SFHTS_MetroUS.csv")
df.head()
```

```python
df.shape
```

```python
ZHVI = df.drop(["RegionID", "SizeRank", "RegionType", "StateName"], axis=1)
df = df.reset_index(drop=True)
ZHVI = ZHVI.transpose()
ZHVI = ZHVI.rename(columns=ZHVI.iloc[0]).drop(ZHVI.index[0])
ZHVI.index.name = 'Time'
ZHVI.head()
```

```python
pct_change_df = ZHVI.pct_change()
pct_change_df = pct_change_df.set_index(ZHVI.index)
pct_change_df.head()
```

```python
print(pct_change_df.shape)
```

```python
avg_row = pct_change_df.mean(axis=0)
# The average ZHVI of United States
print(avg_row["United States"])
sorted_avg_row = avg_row.sort_values(ascending=False)
print(sorted_avg_row.head(20))
# These are the twenty cities have the highest ZHVI (Zillow Housing Value Index) for the Single Family Housing
top20_ZHVI = sorted_avg_row.head(20).index.tolist()
print(type(avg_row))
```

```python
df_ZHVF = pd.read_csv("AllHomeValueForecast_MeroUS.csv")
df_ZHVF.head()
```

```python
ZHVF_top20 = df_ZHVF["2024-03-31"].sort_values(ascending = False).head(20)
top20_indices = ZHVF_top20.index.tolist()
display(ZHVF_top20)
ZHVF_top20_regions = df_ZHVF.loc[top20_indices, "RegionName"]
display(ZHVF_top20_regions)
print(type(ZHVF_top20_regions))
```

```python
ZHVF_top_20 = pd.concat([ZHVF_top20_regions, ZHVF_top20], axis=1)
display(ZHVF_top_20)
print(type(avg_row))
```

```python
avg_row.name = "Ave_ZHVI_percen_incre"
ZHVF_top_20 = pd.merge(ZHVF_top_20, avg_row, left_on="RegionName", right_index=True)
ZHVF_top_20 = pd.concat([ZHVF_top_20, df["2023-03-31"]], axis=1, join="inner")
display(ZHVF_top_20)
'''The following locations are forecasted to have the highest increase in home values and 
also have favorable aggregate single-family home values (reflects the typical value for 
homes in the 35th to 65th percentile range) and positive health market changes across the country.'''
```

```python
ZHVF_top_20["2023-03-31"] = ZHVF_top_20["2023-03-31"].apply(lambda x: '${:,.2f}'.format(x))
```
